```python
import pandas as pd

file_path = "/content/drive/MyDrive/Exoplanet_Dataset/koi_data.csv"
df = pd.read_csv(file_path, comment='#')

df.head()
```

```python
df = df[df['koi_disposition'].isin(['CONFIRMED', 'FALSE POSITIVE'])]

df['label'] = df['koi_disposition'].map({
    'CONFIRMED': 1,
    'FALSE POSITIVE': 0
})

df['label'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['koi_disposition'].value_counts().plot(kind='bar')
plt.title("Distribution of Planet Classes")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing.head(20)
```

```python
missing[missing > 0].plot(kind='bar', figsize=(12,5))
plt.title("Missing Values per Column")
plt.show()
```
